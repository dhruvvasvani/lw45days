```python
from sklearn.linear_model import LinearRegression
import numpy as np

sample_df = df[['AccountAge', 'MonthlyCharges']].head(100)
#sample_df = df[['AccountAge', 'MonthlyCharges']].tail(5)

X = sample_df[['AccountAge']]      
y = sample_df['MonthlyCharges']

model = LinearRegression()
model.fit(X, y)

print(" done with 5 point!")
```

```python
input_age = 91
custom_input = np.array([[input_age]])
predicted_charge = model.predict(custom_input)

print(f"🔮 Prediction Results:")
print(f"For an AccountAge of {input_age} months, the predicted MonthlyCharges are: {predicted_charge[0]:.4f}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

custom_data = {
    'hr': [10, 20, 30, 40, 50],         # Change these numbers
    'day': [1,2,3,4,5]  # Change these numbers
}
```

```python
sample_df = pd.DataFrame(custom_data)
X = sample_df[['hr']]
y = sample_df['day']

model = LinearRegression()
model.fit(X, y)

print("done!")
```

```python
import numpy as np
input_hr = 80

custom_input = np.array([[input_hr]])
custom_input_df = pd.DataFrame([[input_hr]], columns=['hr'])
predicted_day = model.predict(custom_input_df)

print(f"Prediction Results:")
print(f"For {input_hr} hours, the predicted day is: {predicted_day[0]:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(sample_df['hr'], y, color='blue', s=100, label='Actual Data Points')

X_line = pd.DataFrame({'hr': np.linspace(0, 60, 100)})

y_line = model.predict(X_line)
plt.plot(X_line['hr'], y_line, color='red', linewidth=2, label='Regression Line')

plt.xlabel('Hours (hr)')
plt.ylabel('Days (day)')
plt.title('Linear Regression Line for Custom Data')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

# Show the final plot
plt.show()
```
