```python
import cv2
import matplotlib.pyplot as plt
image_path = input("Enter your image filename (e.g., my_image.jpg): ")
img = cv2.imread(image_path)

if img is None:
    print("Error: Could not load the image. Check the filename or path!")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    height, width, _ = img_rgb.shape  
    print(f"\nImage Loaded Successfully!")
    print(f"Current Image Dimensions: Width = {width}px, Height = {height}px")
        plt.figure(figsize=(8, 6))
    plt.imshow(img_rgb)
    plt.axis('on')  
    plt.show()
```

```python
print("Enter your target crop box dimensions:")
ymin = int(input(f"Enter TOP boundary (0 to {height}): "))
ymax = int(input(f"Enter BOTTOM boundary (0 to {height}): "))
xmin = int(input(f"Enter LEFT boundary (0 to {width}): "))
xmax = int(input(f"Enter RIGHT boundary (0 to {width}): "))

output_name = input("\nEnter name for the cropped file (e.g., crop_result.png): ")
cropped_img = img_rgb[max(0, ymin):min(height, ymax), max(0, xmin):min(width, xmax)]

if cropped_img.size == 0:
    print("\nError: The crop coordinates resulted in an empty image. Try again!")
else:
    final_bgr = cv2.cvtColor(cropped_img, cv2.COLOR_RGB2BGR)
    cv2.imwrite(output_name, final_bgr)
    print(f"\n🎉 Success! Cropped image saved as '{output_name}'")
    plt.figure(figsize=(5, 4))
    plt.imshow(cropped_img)
    plt.title("Cropped Output Preview")
    plt.axis('off')
    plt.show()
```
