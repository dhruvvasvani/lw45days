```python
import pandas as pd
import numpy as np

print("STEP 1: Simulating Raw Incoming INR Transaction Logs")

raw_inr_data = {
    'InvoiceID': ['INV-001', 'INV-002', 'INV-002', 'INV-004', 'INV-005'],  
    'Amount_Raw': ['₹1,500', 'Rs. 450', 'Rs. 450', '₹-200', '₹3,200'],    
    'PaymentMode': ['UPI', 'upi', 'Net Banking', 'Cash', np.nan],         
    'CustomerCity': ['Mumbai', 'Delhi', 'Delhi', 'Bangalore', 'Chennai']
}
df_inr = pd.DataFrame(raw_inr_data)
print("Raw INR Data Frame successfully initialized:")
df_inr
```

```python
print("STEP 2: Structural Inspection Diagnostics")

print(f"Dataset Shape: {df_inr.shape}")
print(f"Duplicate Invoices Detected: {df_inr.duplicated(subset=['InvoiceID']).sum()}")
print("\nMissing Values Count:")
print(df_inr.isnull().sum())
```

```python
print("STEP 3: Deduplication and Currency Normalization")

df_inr_clean = df_inr.drop_duplicates(subset=['InvoiceID'], keep='first').reset_index(drop=True)

df_inr_clean['Amount_INR'] = df_inr_clean['Amount_Raw'].str.replace('₹', '', regex=False)
df_inr_clean['Amount_INR'] = df_inr_clean['Amount_INR'].str.replace('Rs.', '', regex=False)
df_inr_clean['Amount_INR'] = df_inr_clean['Amount_INR'].str.replace(',', '', regex=False)
df_inr_clean['Amount_INR'] = df_inr_clean['Amount_INR'].str.strip()

df_inr_clean['Amount_INR'] = pd.to_numeric(df_inr_clean['Amount_INR'])

df_inr_clean[['InvoiceID', 'Amount_Raw', 'Amount_INR']]
```

```python
print("STEP 4: Handling Anomalies, Missing Modes & Text Formatting")
df_inr_clean['PaymentMode'] = df_inr_clean['PaymentMode'].replace({'upi': 'UPI'})
frequent_mode = df_inr_clean['PaymentMode'].mode()[0]
df_inr_clean['PaymentMode'] = df_inr_clean['PaymentMode'].fillna(frequent_mode)

df_inr_clean['Amount_INR'] = df_inr_clean['Amount_INR'].abs()

df_inr_final = df_inr_clean.drop(columns=['Amount_Raw'])
print("Final Cleaned, Analysis-Ready INR Dataset:")
df_inr_final
```
