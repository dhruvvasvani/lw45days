```python
# Project 5: Statistical Decision Support
# Framework: Validating Business Assumptions with Empirical Testing
```

```python
import pandas as pd
import scipy.stats as stats

# Daily store revenue (in ₹) over 7 days for the Control Group (No Free Shipping Offer)
control_revenue = [45000, 48000, 42000, 46000, 44000, 47000, 43000]

# Daily store revenue (in ₹) over 7 days for the Variant Group (Free Shipping over ₹999)
variant_revenue = [52000, 55000, 49000, 53000, 51000, 56000, 50000]

print("Average Daily Revenue (Control Group): ₹", sum(control_revenue) / len(control_revenue))
print("Average Daily Revenue (Variant Group): ₹", sum(variant_revenue) / len(variant_revenue))
```

```python
# 1. Run a Independent Two-Sample T-Test to check if the revenue jump is real
t_stat, p_value = stats.ttest_ind(control_revenue, variant_revenue)

print("Calculated P-Value Result:", round(p_value, 6))
print("-" * 60)

# 2. Apply strict business logic to make the final choice
if p_value < 0.05:
    print("DECISION: GO! The revenue increase is statistically significant.")
    print("Action Plan: Implement the free shipping incentive across 100% of live users.")
else:
    print("DECISION: NO-GO! The difference could be due to pure random variation.")
    print("Action Plan: Reject the change and keep the baseline store model.")
```
