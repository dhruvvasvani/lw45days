```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# 1. Create a simple table of past applicants (The data our AI will learn from)
data = {
    'Credit_Score': [750, 450, 800, 500, 710, 580],
    'Monthly_Income_INR': [90000, 25000, 120000, 30000, 85000, 40000],
    'Loan_Approved': [1, 0, 1, 0, 1, 0]  # 1 means Yes, 0 means No
}

df = pd.DataFrame(data)

# Separate the inputs (X) from the answer column (y)
X_train = df[['Credit_Score', 'Monthly_Income_INR']]
y_train = df['Loan_Approved']

print("--- Past Training Data Table ---")
df
```

```python
# 2. Initialize and teach the AI using our training rows
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

print("AI Training complete! The model has learned the relationship between scores, income, and approvals.")
```

```python
# 3. Define two new applicants walking into the bank today
new_applicants = {
    'Credit_Score': [790, 480],
    'Monthly_Income_INR': [95000, 28000]
}

X_new = pd.DataFrame(new_applicants)

# Ask the AI to predict the outcome
predictions = model.predict(X_new)

# Print out the clear business action steps
for i in range(len(predictions)):
    print(f"Applicant {i+1} (Score: {new_applicants['Credit_Score'][i]}, Income: ₹{new_applicants['Monthly_Income_INR'][i]})")
    if predictions[i] == 1:
        print(" AI DECISION: APPROVE LOAN (Low Risk)")
    else:
        print(" AI DECISION: DENY LOAN (High Risk Default)")
    print("-" * 50)
```
