```python
# Project 2: Python-Based Data Processing Pipeline
# Framework: Applying Business Logic and Transformations

# Business Objective:** Standardize raw customer data, isolate high-risk revenue accounts, and calculate financial exposure to give management an immediate, actionable report on subscription health.

# Applied Business Rules & Transformations:
# 1. Data Sanitization: Handle empty string formatting variations in financial text inputs.
# 2. Feature Optimization: Inject the engineered `ValuePerHour` continuous metric directly into the data tracking frame.
# 3. Risk Stratification Tiering: Classify subscribers into operational categories (`High Risk / High Value`, `High Risk / Low Engagement`, or `Stable Account`) based on behavior thresholds.

# Expected Analytical Output: An executive summary report detailing total at-risk revenue, alongside a processed dataset exported for business intelligence dashboards.
```

```python
import pandas as pd
import numpy as np

print("STEP 1: Simulating Raw Incoming Production Data")

raw_data = {
    'CustomerID': ['USR-101', 'USR-102', 'USR-103', 'USR-104'],
    'MonthlyCharges': [18.99, 7.99, 15.49, 5.00],
    'ViewingHoursPerWeek': [2.5, 35.0, 4.0, 12.0],
    'SupportTicketsPerMonth': [5, 0, 6, 1],
    'TotalCharges': ['45.50', '280.00', ' ', '15.00']
}
df_live = pd.DataFrame(raw_data)
df_live
```

```python
print("STEP 2: Sanitizing Data Formats & Text Fields")
df_live['TotalCharges'] = pd.to_numeric(df_live['TotalCharges'], errors='coerce')

df_live['TotalCharges'] = df_live['TotalCharges'].fillna(0)
print("Data cleaning complete:")
df_live
```

```python
print("STEP 3: Engineering 'ValuePerHour' Metric")
df_live['ValuePerHour'] = df_live['MonthlyCharges'] / (df_live['ViewingHoursPerWeek'] + 0.001)

print("Feature engineered successfully (Isolates users paying high fees with low usage):")
df_live[['CustomerID', 'MonthlyCharges', 'ViewingHoursPerWeek', 'ValuePerHour']]
```

```python
print("STEP 4: Categorizing Accounts via Business Logic Thresholds")

high_value_risk = (df_live['SupportTicketsPerMonth'] > 4) & (df_live['MonthlyCharges'] >= 15.0)
low_engage_risk = (df_live['SupportTicketsPerMonth'] > 4) & (df_live['ViewingHoursPerWeek'] < 10.0)

conditions = [high_value_risk, low_engage_risk]
choices = ['High Risk / High Value', 'High Risk / Low Engagement']

df_live['Operational_Status'] = np.select(conditions, choices, default='Stable Account')

print("Final segmented dataset ready for downstream customer success actions:")
df_live[['CustomerID', 'SupportTicketsPerMonth', 'ViewingHoursPerWeek', 'Operational_Status']]
```
