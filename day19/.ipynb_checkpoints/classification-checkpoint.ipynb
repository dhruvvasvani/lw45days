```python
import pandas as pd

df = pd.read_csv('Titanic-Dataset.csv')
print("Dataset Sample")
print(df.head())
```

```python
gender_survival = df.groupby('Sex')['Survived'].mean()

print("Survival Rate by Gender")
print(gender_survival)
```

```python
gender_counts = df['Sex'].value_counts()

print("Raw Count of Passengers by Gender")
print(gender_counts)
gender_survival_rate = df.groupby('Sex')['Survived'].mean()

print(" survival Rate by Gender")
print(gender_survival_rate)
```

```python
gender_map = {'male': 0, 'female': 1}

df['Sex_encoded'] = df['Sex'].map(gender_map)

print("After Mapping 'Sex' to Numbers")
print(df[['Sex', 'Sex_encoded']].head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(18, 5)) 

sns.barplot(ax=axes[0], x='Sex', y='Survived', data=df, errorbar=None, palette='pastel')
axes[0].set_xlabel('Gender', fontsize=11)
axes[0].set_ylabel('Survival Rate (Percentage)', fontsize=11)
axes[0].set_title('Survival Rate by Gender', fontsize=12, fontweight='bold')

sns.countplot(ax=axes[1], x='Sex', hue='Survived', data=df, palette='Set2')
axes[1].set_xlabel('Gender', fontsize=11)
axes[1].set_ylabel('Passenger Count', fontsize=11)
axes[1].set_title('Body Counts by Gender', fontsize=12, fontweight='bold')
axes[1].legend(title='Outcome', labels=['Died (0)', 'Survived (1)'])

sns.countplot(ax=axes[2], x='Pclass', hue='Survived', data=df, palette='viridis')
axes[2].set_xlabel('Passenger Class', fontsize=11)
axes[2].set_ylabel('Passenger Count', fontsize=11)
axes[2].set_title('Survival Breakdown by Class', fontsize=12, fontweight='bold')
axes[2].legend(title='Outcome', labels=['Died (0)', 'Survived (1)'])

# 2. Automatically adjust the layout so titles and labels don't overlap
plt.tight_layout()

# 3. Show the final merged dashboard
plt.show()
```
