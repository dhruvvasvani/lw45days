```python
import cv2
import matplotlib.pyplot as plt

img1 = cv2.imread('download.jpg')
img2 = cv2.imread('download.jpg')

img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

size_small = (200, 200)
img1_resized = cv2.resize(img1_rgb, size_small)

size_wide = (600, 300)
img2_resized = cv2.resize(img2_rgb, size_wide)

print(f"Variable 'img1_resized' shape: {img1_resized.shape}")
print(f"Variable 'img2_resized' shape: {img2_resized.shape}\n")

plt.figure(figsize=(8, 4))
plt.imshow(img1_resized)
plt.title(f"Image 1 (Small Square): {size_small[0]}x{size_small[1]}")
plt.axis('off')
plt.show()

plt.figure(figsize=(8, 4))
plt.imshow(img2_resized)
plt.title(f"Image 2 (Wide Banner): {size_wide[0]}x{size_wide[1]}")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

flag = np.zeros((400, 600, 3), dtype=np.uint8)

flag[0:133, :] = [255, 153, 51]

flag[133:266, :] = [255, 255, 255]

flag[266:400, :] = [19, 136, 8]


center = (300, 200) # Exact middle of our 600x400 canvas
navy_blue = (0, 0, 128)
cv2.circle(flag, center, 45, navy_blue, 3)
cv2.circle(flag, center, 4, navy_blue, -1)

import math
for i in range(24):
    angle = math.radians(i * 15)
    
    end_x = int(center[0] + 45 * math.cos(angle))
    end_y = int(center[1] + 45 * math.sin(angle))
    
    cv2.line(flag, center, (end_x, end_y), navy_blue, 2)


plt.figure(figsize=(9, 6))
plt.imshow(flag)
plt.axis('off')
plt.title("National Flag of India", fontsize=14, fontweight='bold')
plt.show()
```
