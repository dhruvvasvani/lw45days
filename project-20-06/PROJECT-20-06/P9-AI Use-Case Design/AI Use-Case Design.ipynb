```python
#Simulating the NLP Feature Vector Engine
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

#ref std job des

job_description = ["Looking for an IT Engineering graduate with skills in Python, Machine Learning, SQL database management, and full-stack development."]

#raw resume data

incoming_resumes = {
    'Applicant_ID': ['APP-801', 'APP-802', 'APP-803'],
    'Resume_Text': [
        "IT student specialized in Python, Machine Learning models, and SQL backend databases.", 
        "Experienced sales manager with a history of driving retail shoe store revenue and team leadership.", 
        "Full-stack software developer focusing on web application design, JavaScript, Node.js, and Express." 
    ]
}
#nlp vectorization, converts raw txt into mthmtl wght
df_applicants = pd.DataFrame(incoming_resumes)
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
all_texts = job_description + list(df_applicants['Resume_Text'])
tfidf_matrix = tfidf_vectorizer.fit_transform(all_texts)

#cosine similarity score between the job description and each resume
scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])
df_applicants['Match_Score_Percentage'] = (scores[0] * 100).round(2)
print("--- AI Parsing & Ranking Output ---")
df_applicants.sort_values(by='Match_Score_Percentage', ascending=False)
```
