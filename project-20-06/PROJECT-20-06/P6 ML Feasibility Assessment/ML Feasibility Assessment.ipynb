```python
import pandas as pd

project_pipeline = {
    'Project_Name': ['Automated_Refund_Lookup', 'Customer_Churn_Prediction', 'High_Value_Shopper_Clustering'],
    'Dataset_Rows': [50000, 450, 12000],
    'Has_Clean_Labels': [True, True, False],
    'Needs_Complex_Patterns': [False, True, True] 
}

df_pipeline = pd.DataFrame(project_pipeline)
print("Corporate Project Pipeline Awaiting Vetting")
df_pipeline
```

```python
print("--- Running Machine Learning Feasibility Diagnostics ---")

def evaluate_ml_feasibility(row):
    if not row['Needs_Complex_Patterns']:
        return "REJECT (Use Simple Database/SQL Logic Instead)"
    
    if row['Dataset_Rows'] < 1000:
        return "REJECT (Data Volume Too Low - Missing Statistical Scale)"
    
    if row['Has_Clean_Labels']:
        return "FEASIBLE (Approach: Supervised Learning / Classification)"
    else:
        return "FEASIBLE (Approach: Unsupervised Learning / K-Means Clustering)"

df_pipeline['Feasibility_Assessment'] = df_pipeline.apply(evaluate_ml_feasibility, axis=1)

df_pipeline[['Project_Name', 'Feasibility_Assessment']]
```
